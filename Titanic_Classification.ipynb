```python
clf1 = setup(data = train, 
             target = 'Survived',
             numeric_imputation = 'mean',
             categorical_features = ['Sex','Embarked'], 
             ignore_features = ['Name','Ticket','Cabin'],
             silent = True)
```

```python
gbc = create_model('gbc')
```

```python
tuned_gbc = tune_model(gbc)
```

```python
plot_model(estimator = tuned_gbc, plot = 'learning')
```
