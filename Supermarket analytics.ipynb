```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import sem
import seaborn as sns

df = pd.read_csv('/Users/boss/Downloads/supermarket_sales - Sheet1.csv')
df.head()
```

```python
sns.barplot(x = 'City', y = 'Unit price', data = df, hue = 'Payment',ci=0)
```

```python
sns.scatterplot(x = 'Quantity', y = 'Unit price', data = df, hue = 'City')
```

```python
sns.scatterplot(x = 'Date', y = 'cogs', data = df, hue = 'City',alpha = 0.5)
```

```python
df.isnull().sum()
```

```python
np.round(df.corr(),2)

sns.heatmap(np.round(df.corr(),2),annot = True)
```

```python
plt.figure(figsize=(12,6),dpi=100)
sns.regplot(x='Tax 5%',y='gross income',data=df,color='Red')
plt.xlabel('Tax 5%')
plt.ylabel('Gross Income')
plt.show()
```

```python
sns.countplot(df['Branch'])
plt.xlabel('Branch Name')
plt.ylabel('Count')
plt.title('Which Branch is the most busy?')
```

```python
sns.catplot(x='Product line',y='Unit price',hue='City',kind='swarm',data=df,aspect=2)
plt.xlabel('Product Type')
plt.ylabel('Unit Price')
```
